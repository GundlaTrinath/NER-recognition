```python
import spacy
from spacy import displacy
```

```python
text_file = open(r'Outputs/Clean-Tokens.txt',"r")
```

```python
print(text_file)
```

```python
with text_file as f:
    content = f.read()
```

```python
print(content)
```

```python
trf = spacy.load("en_core_web_trf")
```

```python
doc = trf(content)
```

```python
displacy.render(doc, style='ent')
```

```python
spacy.explain('NORP')
```
